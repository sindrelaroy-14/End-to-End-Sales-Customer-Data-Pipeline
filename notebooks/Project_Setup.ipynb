```python
import json
import os

# 1. Define the schema content
schema_dict = {
  "customers": "customer_id STRING, customer_unique_id STRING, customer_zip_code_prefix INT, customer_city STRING, customer_state STRING",
  "orders": "order_id STRING, customer_id STRING, order_status STRING, order_purchase_timestamp TIMESTAMP, order_approved_at TIMESTAMP, order_delivered_carrier_date TIMESTAMP, order_delivered_customer_date TIMESTAMP, order_estimated_delivery_date TIMESTAMP",
  "order_items": "order_id STRING, order_item_id INT, product_id STRING, seller_id STRING, shipping_limit_date TIMESTAMP, price DOUBLE, freight_value DOUBLE",
  "products": "product_id STRING, product_category_name STRING, product_name_lenght INT, product_description_lenght INT, product_photos_qty INT, product_weight_g INT, product_length_cm INT, product_height_cm INT, product_width_cm INT"
}

# 2. Define a path in your Workspace (Not DBFS)
# This path is 'file-like' and works in new Free Edition
workspace_path = "/Workspace/Users/roysindrela@gmail.com/End-to-End-Sales-Customer-Data-Pipeline/data/"
os.makedirs(workspace_path, exist_ok=True)

# 3. Save the JSON file using standard Python 'open'
file_full_path = os.path.join(workspace_path, "schema_definitions.json")
with open(file_full_path, "w") as f:
    json.dump(schema_dict, f)

print(f"Success! Schema saved at: {file_full_path}")
```
